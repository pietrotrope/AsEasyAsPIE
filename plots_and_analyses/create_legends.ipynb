```python
import matplotlib.pyplot as plt 
import matplotlib.lines as mlines
width = 10
height = 3

prop_cycle = plt.rcParams['axes.prop_cycle']
colors_l = prop_cycle.by_key()['color']

fig = plt.figure()

ax = fig.add_subplot(111)
ax.set_axis_off()

plt.figure(figsize=(width, height))
    

blue = mlines.Line2D([], [], color=colors_l[0], marker='s', linestyle='None',
                                    markersize=100, label='All samples')
orange = mlines.Line2D([], [], color=colors_l[1], marker='s', linestyle='None',
                        markersize=100, label='PIEs')

dotted = mlines.Line2D([], [], color='black', linestyle='--',
                        markersize=100, label='Pruned models',linewidth=10)

not_dotted = mlines.Line2D([], [], color='black', linestyle='-',
                        markersize=100, label='Unpruned models',linewidth=10)

#plt.legend(handles=)

fig.legend([blue, orange, dotted, not_dotted], ['All samples', 'PIEs', 'Pruned models', 'Unpruned models'], loc="center", bbox_to_anchor=(0.5, 0.5), prop={"size":110}, labelspacing=2, ncol=4)

fig.savefig('legend.pdf', dpi=400, transparent=True, bbox_inches='tight')

    
    
            
        
 



```
